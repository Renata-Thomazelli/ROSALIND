Given: Two positive integers a and b (a<b<10000).

Return: The sum of all odd integers from a through b, inclusively.

I received these integers 4916 and 9583. Let's see the solution

This is a good exercise to learning how to work with loops

```python
#What we want to do? First, we need the sum, soo "soma" which means sum in Portuguese will be my variable. 
#Well I don't know how much will be the sum, that is because a put 0 as value. 
#what I have is the integers range from 1916 through 9583. 
#Soo let's put that in a range
#if a number in my range is indivisible for two I will take this number and I'll sum it with the next number that also will be indivisible 
#for two. In the end, I will have the sum of all odd numbers in my range. 
#Important: I can't forget that a also need the last number, but the function range excluded the last number, 
#soo if I need to sum it I need to sum 1 for it.


soma = 0
for i in range (4916,9583+1):
    if i % 2 != 0:
        soma +=i
print (soma)
```
